```python
import json
from IPython.display import display, display_html
from PIL import Image
from analyze import load_model, BATCH_SIZE, load_image_tensor
import torch
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
dataset, loader, model = load_model(device)
num_classes = len(dataset.classes)
```

```python
with open('best10.json') as f:
    json_data = json.load(f)
data = {}
for label, images in json_data.items():
    if len(images) > 0:
        data[label] = images
```

```python
def get_class_name(label):
    return dataset.classes[label]
```

```python
for label, images in data.items():
    display_html(f'<h1>{label}</h1>', raw=True)
    for image in images[:3]:
        path = '../data/processed_images/' + image['image']
        _, tensor = load_image_tensor(path, device=device, unsqueeze=True)
        conf, labels = torch.topk(model(tensor), 3)
        conf, labels = conf[0], labels[0]
        for i in range(3):
            print(get_class_name(labels[i]), conf[i])
        print(image['image'])
        display(Image.open('../data/processed_images/' + image['image']))
```
