```python
%matplotlib inline
import json
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from matplotlib.image import imread

with open('best10.json') as f:
    json_data = json.load(f)

data = {}
for label, images in json_data.items():
    if len(images) > 0:
        data[label] = images


fig, axs = plt.subplots(nrows=len(data), ncols=5, figsize=(12, 35))

for i, (label, images) in enumerate(data.items()):
    for j, image in enumerate(images[:5]):
        if j == 0:
            axs[i][j].set_title(label)
        img = imread('../data/processed_images/' + image['image'])
        axs[i][j].imshow(img)
        axs[i][j].axis('off')

plt.tight_layout()
plt.show()
```
