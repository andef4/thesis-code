```python
from rise import RISE
import matplotlib.pyplot as plt
from testnet import load_dataset
from IPython.core.display import HTML, display
```

```python
train_loader, test_loader = load_dataset(1)

explainer = RISE(None, (240, 240), 10)
explainer.generate_masks(N=10, s=8, p1=0.1)
```

```python
image = next(iter(test_loader))
image = image['input'].squeeze().squeeze().numpy()
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.show()
```

```python
for mask in explainer.masks[:10]:
    display(HTML('<hr>'))
    mask = mask.cpu().numpy()[0]
    plt.imshow(mask)
    plt.colorbar()
    plt.show()
    plt.imshow(image * mask)
    plt.colorbar()
    plt.show()
```
