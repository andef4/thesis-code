```python
from PIL import Image
import numpy as np
from scipy.spatial.distance import directed_hausdorff
from plot_utils import plot_image_row
```

```python
def distance(image1, image2):
    hd1 = directed_hausdorff(image1, image2)[0]
    hd2 = directed_hausdorff(image2, image1)[0]
    return np.max([hd1, hd2])
```

```python
circle = np.array(Image.open('hausdorff/circle.png').convert('L'))
circle_hole = np.array(Image.open('hausdorff/circle_hole.png').convert('L'))
circle_moved_slightly = np.array(Image.open('hausdorff/circle_moved_slightly.png').convert('L'))
circle_moved = np.array(Image.open('hausdorff/circle_moved.png').convert('L'))
circle_moved2 = np.array(Image.open('hausdorff/circle_moved2.png').convert('L'))
circles = np.array(Image.open('hausdorff/circles.png').convert('L'))
square = np.array(Image.open('hausdorff/square.png').convert('L'))
square_moved = np.array(Image.open('hausdorff/square_moved.png').convert('L'))
```

```python
def plot(image1, image2, name):
    dst = distance(image1, image2)
    plot_image_row([image1, image2], title=f'{name}: {dst:.2f}')
```

```python
plot(circle, circle, 'Unchanged')
plot(circle, circle_hole, 'Hole')
plot(circle, circle_moved_slightly, 'Moved slighty')
plot(circle, circle_moved, 'Moved halfway')
plot(circle, circle_moved2, 'Moved to right')
plot(circle, square, 'Square')
plot(circle, square_moved, 'Moved square')
plot(circle, circles, 'Multiple circles')
```
