```python
from PIL import Image
from pathlib import Path
import numpy as np
from plot_utils import plot_image_row
```

```python
sample = 'Brats18_2013_0_1'
base_path = Path('data/processed') / sample


def open_image(ct_slice, filename):
    return np.array(Image.open(base_path / ct_slice / filename))


for ct_slice in ['L1', 'L2', 'L3']:
    plot_image_row([
        open_image(ct_slice, 't1.png'),
        open_image(ct_slice, 't1ce.png'),
        open_image(ct_slice, 't2.png'),
        open_image(ct_slice, 'flair.png'),
        open_image(ct_slice, 'segment.png'),
        ],
        title=f'{sample} {ct_slice}',
        labels=['T1', 'T1ce', 'T2', 'Flair', 'Segment'],
        color_maps=['gray', 'gray', 'gray', 'gray', 'viridis'],
    )
```
